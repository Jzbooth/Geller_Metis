```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
y  = pd.DataFrame(np.random.rand(200))
X = pd.DataFrame(np.random.rand(200,20))
```

```python
model = sm.OLS(y,X)
fit = model.fit()
fit.summary()
```

```python
X = pd.DataFrame(np.random.rand(200,40))
model = sm.OLS(y,X)
fit = model.fit()
fit.summary()
```

```python
X = pd.DataFrame(np.random.rand(200,60))
model = sm.OLS(y,X)
fit = model.fit()
print(fit.pvalues)
```
