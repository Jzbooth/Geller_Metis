````python
def sel_words(str1,str2):
    str1=str1.split()
    str2=str2.split()
    mmg_more_four=[]
    geb_more_four=[]
    for w in str1:
        if len(w)>=5:
            mmg_more_four.append(w)
    for w in str2:
        if len(w)>=5:
            geb_more_four.append(w)
    return mmg_more_four,geb_more_four

print(sel_words(mmg,geb))
````

````python
import itertools
````

````python
def find_intersection(list1,list2):
    intersection = []
    for word1 in list1:
        for word2 in list2:
            if word1 == word2:
                intersection.append(word1)
    return intersection
````

````python
list1 = ['fhdksl', 'djiw2', 'fjdis', 'fyeiow','ghfjs']
list2 = ['fidhwie','djiw2','fyeiow','dkfjwodk','lkdfjie','ghfjs']
````

````python
find_intersection(list1,list2)
````

````python
def make_three_word_phrase(intersection):
    if len(intersection) > 2:
        phrases = []
        comb = itertools.combinations(list1, 3)
        for c in comb:
            phrases.append(str(c[0]+ ' '+c[1]+ ' ' + c[2]))
        return phrases
    else:
        return 'There are less than 3 words'
````

````python
intersection = find_intersection(list1, list2)
make_three_word_phrase(intersection)
````
