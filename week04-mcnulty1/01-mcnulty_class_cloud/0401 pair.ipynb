```python
#Olando's:
def brackets(s):
    openers = ["(", "{", "["]
    closers = {"}":"{", "]":"[", ")":"("}
    stack =[]
    s = s.strip()
    for i in s:
        if i in openers:
            stack.append(i)
        else:
            try:
                if stack.pop() != closers[i]:
                    # something wasn't paired
                    return False
            except:
                # there was a surplus of closers
                return False
    if stack:
        # there was a surplus of openers
        return False
    # no fail condidtions met; string is balanced
    return True
```
